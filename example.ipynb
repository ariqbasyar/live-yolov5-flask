```python
import cv2
import sys

from models.experimental import attempt_load
from model import detect, box_label, preprocess, get_device
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
cam = cv2.VideoCapture(0)
```

```python
img = cv2.imread('dataset/zidane.jpg')

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
device = get_device()
print(f'using {device}')
preprocessed_img = preprocess(img, device, size=(1280,640))
```

```python
modelv5s = attempt_load('weights/yolov5s.pt', map_location=device)
```

```python
pred = detect(modelv5s, preprocessed_img)
```

```python
labeled_img = box_label(pred, img.copy())
```

```python
plt.imshow(cv2.cvtColor(labeled_img, cv2.COLOR_BGR2RGB))
plt.show()
```
